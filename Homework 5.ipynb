### B. X is with mean μ = 40, Find and plot the density distribution with three different standard deviations (2,6,12) (Please draw the curves on the same figure with different colors and a legend) for:

a. P(x < 40)<br><br>
b. P(x > 21)

```r
pnorm(40, 40, 2)
pnorm(40, 40, 6)
pnorm(40, 40, 12)
```

```r
plot(dnorm(0:40,40,2),type = "l")
lines(dnorm(0:40, 40, 6), lty = 2, lwd = 2, col = "red")
lines(dnorm(0:40, 40, 12), lty = 3, lwd = 2, col = "blue")
legend("topleft",
      legend = expression(italic("SD = 2"),
                         italic("SD = 6"),
                         italic("SD = 12")),
      lty = c(1,2,3),
      col = c("black", "red", "blue"),
      lwd = c(5,2,2))
```

```r
pnorm(21, 40, 2, lower.tail = F)
pnorm(21, 40, 6, lower.tail = F)
pnorm(21, 40, 12, lower.tail = F)
```
