1. The data in the table give the infant mortality rates per 1000 livebirths in the United States for the period 1960-2000


![title](HW1.png)

a. Fit a linear regression line relating infant mortality rate to chronological year using these data. Use a data transformation if necessary

b. Test for significance of the linear relationship

c. If the present trend continues for the next 5 years what would be the predicted infant mortality rate in 2010


```r
#a
x <- c(1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005)
y <- c(26.0, 24.7, 20.0, 16.1, 12.6, 10.6, 9.2, 7.6, 6.9, 6.9)

mod1 <- lm(y ~ x)
plot(x, y, xaxt="n")
axis(side=1, at = seq(1960, 2005, by = 5))
axis(side=2, at= seq(6.5,26.0,0.5))
abline(mod1, lwd=2)
grid(nx = NULL,ny = NULL)
```

```r
#b
# calculate residuals and predicted values
plot(x, y, xlim=c(min(x)-5, max(x)+5), ylim=c(min(y)-10, max(y)+10))
abline(mod1, lwd=2)
res <- signif(residuals(mod1), 5)
pre <- predict(mod1) # plot distances between points and the regression line
segments(x, y, x, pre, col="red")

# add labels (residual values) to points
install.packages("calibrate")
library(calibrate)
textxy(x, y, res, cex=0.7)
```

```r
cor_res <- cor.test(y, x, 
                method = "pearson")
cor_res

#There is high correlation, negative, with infant mortality rate
#and year.  Meaning that the evidence suggested in this data is
#that infant mortality is negatively correlated with time; or
#as time goes on, infant mortality decreases.
```

```r
#c
mod <- lm(x ~ y) 
coef(mod)

#In year 2010, we would expect a negative infant mortality rate;
#which is impossible.  I would expect that near 2010 we are 
#nearing the plateau or threshold and would expect the
#mortality rate to level off around the 6.9 rate as occurred
# in 2000 and 2005.
```

```r
Hacker <- read.csv("HackerRank-Developer-Survey-2018-Numeric.csv", header = T)
head(Hacker)
```

```r
#install.packages("NHANES")
#install.packages("mosaic")
#install.packages("swirl")
library(dplyr)
library(NHANES)
library(ggplot2)
library(mosaic)
library(swirl)
```

```r
Hacker <- Hacker %>% 
mutate(q3Gender = as.numeric(q3Gender == "1", "2"))
```

```r
log_plot <- ggplot(data = Hacker, aes(x = q1AgeBeginCoding, y = q3Gender)) + 
geom_jitter(alpha = 0.1, height = 0.05) + 
geom_smooth(method = "glm", method.args = list(family = "binomial")) + 
ylab("Gender")

log_plot + xlab("Age Began Coding")
```

```r
log_plot + aes(x = q2Age) + xlab("Age")
```

```r
#a
logreg <- glm(q3Gender ~ q1AgeBeginCoding + q2Age, family = "binomial", data = Hacker) 
msummary(logreg)
```

```r
#b
mod_parallel <- lm(q8JobLevel ~ q1AgeBeginCoding + q3Gender, data = Hacker) 
coef(mod_parallel)
rsquared(mod_parallel)
```

```r
plotModel(mod_parallel, system = "ggplot2")
```

```r
mod_interact <- lm(q8JobLevel ~ q1AgeBeginCoding + q3Gender * q1AgeBeginCoding, data = Hacker)
coef(mod_interact)
rsquared(mod_interact)

#It does not appear that age to begin coding, or gender have
#much effect on job level for women.  So my advice would be
#if you want to start coding and you are a woman, you should
#start now, at any age.
```

$3.$ The data set below represents the expected recovery time (in weeks) for patients after a surgical procedure with different post-operational methods for recovery. Negative values represent a faster recovery time than expected. The control patients followed a rutine post-operational for this type of procedure.

Using the data set below compare the effect of three treatments on the expected time of recovery after surgery. (Use ANOVA)

```r
Recovery = c(.53, .36, .20, -.37, -.60, -.64, -.68, -1.27, .73, .31, .03, -.29, -.56, -.96, -1.61,
         -.78, -.86, -1.35, -1.48, -1.52, -2.04, -2.83)
treatment = c(rep("control",8), rep("surgery",7), rep("acupunture",7))
data = data.frame(Recovery, treatment)
```

```r
set.seed(1234)
dplyr::sample_n(data, 10)
```

```r
levels(data$treatment)
```

```r
data$treatment <- ordered(data$treatment,
                         levels = c("control", "acupunture", "surgery"))
```

```r
group_by(data, treatment) %>%
  summarise(
    count = n(),
    mean = mean(Recovery, na.rm = TRUE),
    sd = sd(Recovery, na.rm = TRUE)
  )
```

```r
#install.packages("ggpubr")
library("ggpubr")
ggboxplot(data, x = "treatment", y = "Recovery", 
          color = "treatment", palette = c("#00AFBB", "#E7B800", "#FC4E07"),
          order = c("control", "acupunture", "surgery"),
          ylab = "Recovery", xlab = "Treatment")
```

```r
#library("ggpubr")
ggline(data, x = "treatment", y = "Recovery", 
       add = c("mean_se", "jitter"), 
       order = c("control", "acupunture", "surgery"),
       ylab = "Recovery", xlab = "Treatment")
```

```r
# Box plot
boxplot(Recovery ~ treatment, data = data,
        xlab = "Treatment", ylab = "Recovery",
        frame = FALSE, col = c("#00AFBB", "#E7B800", "#FC4E07"))
# plotmeans
library("gplots")
plotmeans(Recovery ~ treatment, data = data, frame = FALSE,
          xlab = "Treatment", ylab = "Recovery",
          main="Mean Plot with 95% CI") 
```

```r
# Compute the analysis of variance
res.aov <- aov(Recovery ~ treatment, data = data)
# Summary of the analysis
summary(res.aov)
```

```r
TukeyHSD(res.aov)
```

```r
# 1. Homogeneity of variances
plot(res.aov, 1)
```

```r
library(car)
leveneTest(Recovery ~ treatment, data = data)
```

```r
# 2. Normality
plot(res.aov, 2)
```

```r
# Extract the residuals
aov_residuals <- residuals(object = res.aov )
# Run Shapiro-Wilk test
shapiro.test(x = aov_residuals )
```
